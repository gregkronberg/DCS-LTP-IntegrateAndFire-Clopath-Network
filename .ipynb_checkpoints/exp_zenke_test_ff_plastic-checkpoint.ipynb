```python
data_directory = 'Data/'+exp.experiment_name+'/'
trial_id = exp._generate_trial_id()
keyword='_'.join(['t_H', str(exp.P.neurons['E_rate']['t_H']), 'gamma', str(exp.P.neurons['E_rate']['t_H']), 'IE', str(w_IE_init), 'FE', str(w_FE_init) ])
trial_id=trial_id+keyword
file_name_W = 'W'+'_'+trial_id
file_name_W_cons = 'W_cons'+'_'+trial_id
exp.W = {}
exp.W_cons = {}
#-------------------------------------------------
exp.W['E_input_E'] = exp._to_weight_matrix(synapse_group=exp.synapses['E_input_E'], w_key='w_exc_plastic')
exp.W_cons['E_input_E'] = exp._to_weight_matrix(synapse_group=exp.synapses['E_input_E'], w_key='w_cons')
exp.W['EE'] = exp._to_weight_matrix(synapse_group=exp.synapses['EE'], w_key='w_exc_plastic')
exp.W_cons['EE'] = exp._to_weight_matrix(synapse_group=exp.synapses['EE'], w_key='w_cons')
exp.W['IE'] = exp._to_weight_matrix(synapse_group=exp.synapses['IE'], w_key='w_inh_plastic')
exp.W['II'] = exp._to_weight_matrix(synapse_group=exp.synapses['II'], w_key='w_inh_plastic')
exp.W['EI'] = exp._to_weight_matrix(synapse_group=exp.synapses['EI'], w_key='w_exc_plastic_init')
exp.W['sort_by_receptive_field_loc'] = exp.sort_by_receptive_field_loc

exp._save_data(data=exp.W, file_name=file_name_W, data_directory=data_directory)
exp._save_data(data=exp.W_cons, file_name=file_name_W_cons, data_directory=data_directory)
```
